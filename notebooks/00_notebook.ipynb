## Jupyter Components: Server, Dashboard, and Kernels

When you launched this notebook, you encountered two key Jupyter components: the notebook *server* and *dashboard* interface.

We started by launching a notebook server in the terminal:

```bash
jupyter notebook
```

This prints log messages similar to:

```
$ jupyter notebook
[I 08:58:24.417 NotebookApp] Serving notebooks from local directory: ~/mlpractical
[I 08:58:24.417 NotebookApp] 0 active kernels
[I 08:58:24.417 NotebookApp] The Jupyter Notebook is running at: http://localhost:8888/
```

The last message shows the URL where the application is served. By default, `jupyter notebook` opens a browser tab pointing to this address. To launch without opening a browser (useful for remote machines over SSH), use:

```bash
jupyter notebook --no-browser
```

For other command options, see:

```bash
jupyter notebook --help
```

While running, the server continues printing log messages to the terminal. Keep the terminal session open to maintain the server. To shut down the server, press `Ctrl+C` (confirm with `y` or press `Ctrl+C` again to skip confirmation).

### The Dashboard

When Jupyter opens in your browser, you see the notebook *dashboard*:

<img src='res/jupyter-dashboard.png' />

The dashboard shows the `Files` tab - a file browser for the directory where you launched the server. You can:
- Navigate into subdirectories by clicking folder names
- Go to parent directories by clicking `..`
- Create new notebooks using the `New` dropdown
- Perform file management (rename, delete) by selecting files with checkboxes

**Security note:** You can only access files at or below the directory where you started the server.

The `Running` tab shows currently active notebook instances, useful for tracking and managing running notebooks.

## The Notebook Interface

The top of your notebook looks like this:

<img src='res/jupyter-notebook-interface.png' />

Key interface elements:

**Notebook name:** Click to edit the filename shown at the top.

**Checkpoint status:** Shows the last manual save point. Auto-saves happen regularly; checkpoints help recover from unwanted changes.

**Menus:**
- `File`: Save, create new notebooks, copy current notebook, export to various formats
- `Edit`: Standard clipboard functions, cell organization (move, merge, split, insert)
- `Cell`: Run code cells, change cell types
- `Kernel`: Manage the Python process - interrupt busy kernels, restart (clears all variables)

**Kernel status indicator:** Far right of the toolbar
- Dark filled circle: Kernel busy (commands queued)
- Open circle: Kernel idle

**Toolbar:** Quick access to common commands (clipboard, cell/kernel management)

For more details, try the `User Interface Tour` under the `Help` menu.

## Code Cells: Interactive Code Execution

Here's an executable code cell. To run it:
1. Click the cell to select it
2. Press the ▶️ button in the toolbar, use `Cell > Run Cells`, or press `Ctrl+Enter`

````python
import sys

print('Hello world!')
print('Alarming hello!', file=sys.stderr)
print('Hello again!')
'And again!'
````

This example demonstrates the three main components of a code cell:

### Input Area
The code input area automatically provides syntax highlighting for the code you enter.

### Execution Counter
To the left of the input area is the execution indicator:
- `In [ ]:` before the cell is first run
- `In [n]:` after execution, where `n` is the execution counter (incremented for each cell run)

The counter helps track the order cells were executed. While cells don't need to be run in order, keeping them organized sequentially improves readability.

### Output Area
Below the input area, showing:

1. **Standard output (`stdout`):** Regular print statements
2. **Standard error (`stderr`):** Error messages and warnings  
3. **Display area:** The return value of the last Python statement (like in a Python interpreter)

**Note:** All `stdout` output appears together, even if `stderr` output occurred between statements, which explains the unusual ordering above.

## Useful Keyboard Shortcuts

Jupyter has two modes with different shortcuts:

**Command mode** (blue cell border): No cell is being edited
**Edit mode** (green cell border): Currently editing a cell content

### Run Cell Shortcuts (both modes):
- **`Ctrl+Enter`**: Run selected cell
- **`Shift+Enter`**: Run cell and select next
- **`Alt+Enter`**: Run cell and insert new cell below

### Command Mode Shortcuts:
- **`L`**: Toggle line numbers for the selected cell

For a complete list, see `Help > Keyboard Shortcuts`.

## Magic Commands

IPython provides *magic* commands for helpful functionality beyond standard Python. Here are three particularly useful ones for this course:

- **`%%timeit`**: Time cell execution and show statistics (place at cell start)
- **`%precision`**: Set display precision for floating-point numbers and NumPy arrays
- **`%debug`**: Activate interactive debugger after an exception occurs

For a full list, see the [IPython magic commands documentation](http://ipython.readthedocs.io/en/stable/interactive/magics.html).

## Plotting with Matplotlib

When using `matplotlib` in notebooks, first run:

```python
%matplotlib inline
```

This displays all plots as images in the cell output area. Below is an example showing basic plotting options:

````python
# Enable inline plotting
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create plot
fig, ax = plt.subplots(figsize=(8, 4))

# Plot lines with labels
ax.plot(x, y1, linewidth=2, label=r'$\sin(x)$')
ax.plot(x, y2, linewidth=2, label=r'$\cos(x)$')

# Customize plot
ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$y$', fontsize=14)
ax.set_xlim(0, 2 * np.pi)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
